```python
import json
import pandas as pd
import numpy as np
import codecs #For read xml data
import os #Read file in system
import statsmodels.api as sm
```

```python
#Load the file and read as excel document
xls = pd.ExcelFile('aihw-injcat-213-Data_tables_E.xlsx')
df1 = pd.read_excel(xls, 'E1 Numbers')
#Remove empty rows
df1 = df1.dropna(how='all', axis=1)
#Remove last four rows and other no important rows
df1.drop(df1.tail(4).index,inplace=True)
df1.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8'], axis=1,inplace=True)
#grab the first row for the header
new_header = df1.iloc[1] 
#take the data less the header row
df1 = df1[2:] 
#set the header row as the df header
df1.columns = new_header
# Change the year column and only keep last four digits
df1['Year'] = [x[:4] for x in df1['Year']] 
df1
```

```python
#Read the html table
df2 = pd.read_html('table5.html')
df2 = df2[0]
#Keep columns only useful
cols = [3,4,5,6,7,8,9,10,13,14]
df2.drop(df2.columns[cols],axis=1,inplace=True)
#Change columns titles
df2 = df2[1:] 
df2.columns = ['injury_type', 'world_deaths', 'world_rate', 'searo_deaths', 'searo_rate'] 
df2
```

```python
#Output files as xlsx
df1.to_excel('ingury_in_years.xlsx', index = False)
df2.to_excel('ingury_in_world.xlsx', index = False)
```
