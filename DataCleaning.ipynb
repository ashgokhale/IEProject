```python
import pandas as pd
```

```python
inj_years = pd.read_csv('ingury_in_years.csv')
```

```python
inj_years['Year'] = inj_years['Year'].apply(lambda x: x[:4])
```

```python
inj_world = pd.read_csv('ingury_in_world.csv')
inj_world
```

```python
inj_world['World Deaths'] = inj_world['World Deaths'].apply(lambda x: round(x))
inj_world['Australia Deaths'] = inj_world['Australia Deaths'].apply(lambda x: round(x))
inj_world
```
